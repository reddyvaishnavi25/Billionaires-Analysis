```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Billionaire.csv")
data.head(5)
```

```python
result=data.head(10)
print(result)
```

```python
print(data.isnull().sum())
```

```python
data=data.dropna()
```

```python
print(data.isnull().sum())
```

```python
data["NetWorth"]=data["NetWorth"].str.strip("$")
data["NetWorth"]=data["NetWorth"].str.strip("B")
```

```python
print(data.head(10))
```

```python
data["NetWorth"]=data["NetWorth"].astype(float)
```

```python
print(data.head(5))
```

```python
df=data.sort_values(by=["NetWorth"],ascending=False).head(10)
```

```python
print(df)
```

```python
plt.figure(figsize=(9,5))
plt.bar(df["Name"],df["NetWorth"])
plt.xticks(rotation=45)
plt.title("According to the NetWorth")
plt.show()
```

# top 5 domains with the most number of Billionaires

```python
a=data["Source"].value_counts().head(5)
```

```python
print(a)
```

```python
index=a.index
print(a.index)
val=a.values
print(a.values)
```

```python
col=["red","yellow","green","orange","black"]
plt.figure(figsize=(4,4))
plt.pie(val,labels=index,colors=col,shadow=True)
center=plt.Circle((0,0),0.5,color='white')
fig=plt.gcf()
fig.gca().add_artist(center)
plt.rc('font',size=12)
plt.title("top 5 domains with the most number of billionaires",fontsize=10)
plt.show()
```
